```python
import math

pie_data = {
    "Apples": 30,
    "Bananas": 20,
    "Cherries": 15,
    "Grapes": 25,
    "Oranges": 10
}
for i in range(1, 100, 1):
    f = lambda x: math.exp(math.sin(x**2 * i/10))
    g = lambda x: -math.exp(math.sin(x**2 * i / 10))
    plotter = Plotter()
    plotter.plot(f)
    plotter.plot(g)
    plotter.save(f'graph/{i}.png')
```

```python
from HAL.metrika.math import *
from HAL.metrika import Plotter
from HAL.pixels import Color

equation = Equation('log(x)= 0', 'x')

equation.equation

# Differentiate the equation
integral = equation.integrate()
integral2 = integral.integrate()

derivative = equation.differentiate()

equation.display()
derivative.display()
integral.display()
integral2.display()
```

```python
plotter = Plotter()
plotter.draw_axes()
plotter.plot(equation, color = Color.YELLOW)
plotter.plot(integral, color = Color.CYAN)
plotter.plot(derivative, color = Color.RED)
plotter.picture.image
```
